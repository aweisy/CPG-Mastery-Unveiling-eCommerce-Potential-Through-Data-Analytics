```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso



#set random seed before proceeding
np.random.seed(1234)


X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Numeric features for imputation and scaling
numeric_features = ['QUANT_ORDERED', 'UNITARY_SALESPRICEUSD', 'DISCOUNTUSD']  # Add other numeric features as necessary
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

# Categorical features for one-hot encoding
categorical_features = ['DELIVERY_TYPE', 'DEPARTMENT', 'FAMILY']  # Adjust based on your DataFrame

# Combining transformers using ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ])

# Pipeline for Linear Regression with preprocessing
linear_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('linear_regression', LinearRegression())
])

# Pipeline for LASSO with preprocessing and hyperparameter tuning
lasso_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('lasso', Lasso(random_state=42))
])

# Fitting the Linear Regression model
linear_pipeline.fit(X_train, Y_train)
Y_pred_linear = linear_pipeline.predict(X_test)
linear_rmse = root_mean_squared_error(Y_test, Y_pred_linear)
linear_r2 = r2_score(Y_test, Y_pred_linear)

# Setting up Grid Search for LASSO
alpha_grid = {'lasso__alpha': np.logspace(-4, -0.5, 30)}
kf = KFold(n_splits=5, shuffle=True, random_state=42)
grid_search = GridSearchCV(lasso_pipeline, alpha_grid, cv=kf, scoring='neg_mean_squared_error')
grid_search.fit(X_train, Y_train)

# Best LASSO model evaluation
best_alpha = grid_search.best_params_['lasso__alpha']
Y_pred_lasso = grid_search.predict(X_test)
lasso_test_rmse = root_mean_squared_error(Y_test, Y_pred_lasso)
lasso_r2 = r2_score(Y_test, Y_pred_lasso)

print(f"Linear Regression RMSE: {linear_rmse:.2f}")
print(f"Linear Regression R-Squared: {linear_r2:.2f}")
print(f"Best alpha for LASSO: {best_alpha}")
print(f"LASSO Test RMSE with Best Alpha: {lasso_test_rmse:.2f}")
print(f"LASSO Test R-Squared with Best Alpha: {lasso_r2:.2f}")
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam


# Define the model
def build_model(input_shape, output_activation):
    model = Sequential()
    model.add(Dense(2 * input_shape[1], input_dim=input_shape[1], activation='relu'))
    model.add(Dense(1, activation=output_activation))
    return model

# Choose activation function based on outcome type
output_activation = 'sigmoid'  # for categorical outcome
# output_activation = 'relu'  # for numeric outcome

# Build the model
model = build_model(input_shape=X_train.shape, output_activation=output_activation)

# Compile the model
model.compile(optimizer=Adam(), loss='binary_crossentropy' if output_activation == 'sigmoid' else 'mse', metrics=['mae'])

# Train the model
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2)

# Evaluate the model
y_pred = model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("Mean Absolute Error:", mae)
print("Root Mean Squared Error:", rmse)
print("Adjusted R-Squared:", 1 - (1 - r2) * (len(y_test) - 1) / (len(y_test) - X_test.shape[1] - 1))
```
